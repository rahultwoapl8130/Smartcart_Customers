# Analyze The K- value:-- 1.st process use karenge Elbow Method:-- 

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
from kneed import KneeLocator

knee = KneeLocator(range(1, 11), wcss, curve="convex", direction="decreasing")

optimal_k = knee.elbow
```

```python
print("best k= ",optimal_k)
```

```python
# plot
plt.plot(range(1,11), wcss, marker="o")
plt.xlabel("K")
plt.ylabel("WCSS")
```

```python
# Best optimal (K = 4) From The Elbow Method:--
```

```python
# Same Thing We can Figuer Out From The Silhouette Score:--
```

# Silhouette Score:--(2nd process)

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    scores.append(score)

# plot
plt.plot(range(2,11), scores, marker="o")
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs Number of Clusters")
plt.show()
```

```python
# Next Craet Both Plot for Combined The values:--
k_range = range(2, 11)

fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot WCSS
ax1.plot(k_range, wcss[1:], marker="o", color="blue")
ax1.set_xlabel("Number of Clusters (k)")
ax1.set_ylabel("WCSS", color="blue")
ax1.set_title("Elbow Method and Silhouette Score")

# Create second y-axis
ax2 = ax1.twinx()

# Plot Silhouette Score
ax2.plot(k_range, scores[:len(k_range)], marker="x", color="red", linestyle="--")
ax2.set_ylabel("Silhouette Score", color="red")

plt.show()
```

```python
# But yaha par Around (4) Intersect kar Rahe Hii ( Tooo (4) Is the Optimal {K- Value})- Pick for our K-Means:--
```

# ABb hmm kya karneg clustering Algorithm Ko Apply kareneg Isko Aplly Akrne ke Liye :--

```python
# Hmm Sabse Pahle K- Means Ke Sath Work Karte Hii:--
# (K- Means):--
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)

labels_kmeans = kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], 
           c=labels_kmeans, cmap="viridis")

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D PCA Clusters")

plt.show()
```

```python
# to Yaha par (K-means) ne 4(FOUR) Cluster Craet Kar Ke de Diya Hii:--
```

```python
# Aglomarative Clustering
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Create Agglomerative clustering model
agg_cls = AgglomerativeClustering(n_clusters=4, linkage="ward")

# Fit model and get labels
labels_agg = agg_cls.fit_predict(X_pca)

# Create 3D plot
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], 
           c=labels_agg, cmap="viridis")

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("Agglomerative Clustering")

plt.show()
```

```python
# yha par iss Algorithom se better clustering Hua hii :---
# To hmm kya kareneg (Aglomerative clustering Wale Joo  label hii Unhi ke  Sath Hmm next Operation Ko Performe karenge:--
# Kuyki Proper Tarike (FOUR) Well Seperated Cluster De Raha hii:--
# TOo isi Algorithm Ke Upper Hmm Next Operation Ko Performe Karenge:--
# Customer that are existing iN ecah of the clustres :--
# Four Different types ke Customers Cluster de raha hii:--
```
